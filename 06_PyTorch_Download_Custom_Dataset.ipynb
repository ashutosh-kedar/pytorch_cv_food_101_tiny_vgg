```python
from pathlib import Path
import random
root_path = 'data/food-101'

def get_train_test_data(
    root_path=root_path,
    labels=['train','test'],
    target_classes=["pizza", "fried_rice", "ice_cream"],
    amount=0.3):

  data = {}
  for label in labels:
    label_file_path = Path(f'{root_path}/meta/{label}.txt')
    with open(label_file_path,'r') as file:
      label_files = [line.strip('\n') for line in file.readlines() if line.strip('\n').split('/')[0] in target_classes]

    total_samples = round(amount*len(label_files))
    sampled_images = random.sample(label_files,k=total_samples)

    images = [Path(root_path+'/images/'+sample_image+'.jpg') for sample_image in sampled_images]

    data[label] = images
  return data


file_paths = get_train_test_data()
print(f"Total test images: {len(file_paths['test'])}")
print(f"Total train images: {len(file_paths['train'])}")
```

```python
root_structure = 'structured_data/food_data'
path_to_root_structure = Path(root_structure)

if path_to_root_structure.is_dir():
  print(f'{root_structure} directory already exists')
else:
  print(f'Creating {root_structure} directory')
  path_to_root_structure.mkdir(parents=True,exist_ok=True)

```

```python
import shutil


def copy_data_to_structured_folder(
    file_paths=file_paths,
    dest_dir=path_to_root_structure):
  for labels in file_paths.keys():
    for file_path in file_paths[labels]:
      dest_full_path = path_to_root_structure / labels / file_path.parent.stem / file_path.name
      if not dest_full_path.parent.is_dir():
          dest_full_path.parent.mkdir(parents=True,exist_ok=True)
      shutil.copy(file_path,dest_full_path)

copy_data_to_structured_folder()
```

```python
zip_file_name = 'structured_data'
shutil.make_archive(base_name=zip_file_name,
                    root_dir=dest_dir,
                    format='zip')
```
